# 4. Overall Total Satisfaction Level 

**Now we calculate the in general satisfaction level of a customer, ranging from 1-5 (1 lowest, 5 highest) and 0 for not applicable**

**We write a Function such that : <br>**
* **If any of the Rating types(Seat comfort, Inflight wifi service ...) is 0, the Total Satisfaction for that customer would be 0 (not applicable)**
* **Else, we calculate the mean of all the Rating types and round it off to its nearest Integer**

```python
def calculate_satisfaction_level(sc, datc, fd, gl, iws, ie, os, eob, obs, lrs, bh, cs, c, ob):
    list = [sc, datc, fd, gl, iws, ie, os, eob, obs, lrs, bh, cs, c, ob]
    if (sc==0 or datc==0 or fd==0 or gl==0 or iws==0 or ie==0 or os==0 or eob==0 or obs==0 or lrs==0 or bh==0 or cs==0 or c==0 or ob == 0):
        return 0
    else:
        return int(round(np.mean(list)))  
```

**Using the fuction *`calculate_satisfaction_level`* on the dataframe data, and adding a column 'Satisfaction Level'**

```python
data['Satisfaction Level'] = data.apply(lambda data: calculate_satisfaction_level(data['Seat comfort'], data['Departure/Arrival time convenient'], data['Food and drink'], data['Gate location'], data['Inflight wifi service'], data['Inflight entertainment'], data['Online support'], data['Ease of Online booking'], data['On-board service'], data['Leg room service'], data['Baggage handling'], data['Checkin service'], data['Cleanliness'], data['Online boarding']), axis=1)
```
